```python
import module1
import pandas as pd
import numpy as np
from io import StringIO

def read_csv_pandas(csv_filename):
    with open(csv_filename) as f:
        lines = f.readlines()
        return pd.read_csv(StringIO(''.join(lines[3:])), delimiter = ',')

```
